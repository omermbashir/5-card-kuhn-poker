## ðŸš€ Full Run: Find Nash Equilibrium

âš ï¸ **This takes 10-20 minutes.** Run this to find the actual equilibrium.

The solver will:
1. Start with initial strategy guesses
2. Iteratively improve both players' strategies
3. Check for equilibrium every 100 iterations
4. Stop when Îµ < 0.01 (exploitability < 1 cent per hand)


```python
# Full equilibrium solve
print("ðŸŽ¯ Finding Nash Equilibrium")
print("=" * 50)
print("This will take 10-20 minutes...\n")

# Initialize fresh solver
solver_full = EquilibriumSolver(n=5)

# Get starting strategies
p1_start, p2_start = solver_full.strategy_mgr.initialize_default_strategy()

# Run full solver
results_full = solver_full.solve(
    p1_start, p2_start,
    max_iterations=10000,
    epsilon=0.0001,
    check_frequency=100
)

print("\n" + "=" * 50)
if results_full:
    print(f"âœ… Found {len(results_full)} equilibrium!")
    final = results_full[-1]
    print(f"\nðŸ“Š Final Results:")
    print(f"   Iterations: {final['iteration']}")
    print(f"   Player 1 EV: {final['ev_player1']:.6f}")
    print(f"   Player 2 EV: {final['ev_player2']:.6f}")
else:
    print("âš ï¸ Equilibrium not found within iteration limit.")
```

Display the optimal strategies found:

```python
if results_full:
    final = results_full[-1]
    
    print("ðŸŽ¯ PLAYER 1 EQUILIBRIUM STRATEGY")
    print("=" * 60)
    print(final['player1_strategy'].round(3))
    
    print("\nðŸŽ¯ PLAYER 2 EQUILIBRIUM STRATEGY")
    print("=" * 60)
    print(final['player2_strategy'].round(3))
    
    # Interpretation
    print("\nðŸ’¡ INTERPRETATION")
    print("=" * 60)
    
    p1_card3 = final['player1_strategy'].iloc[:, 2]
    print(f"\nPlayer 1 with card 3 (middle):")
    print(f"  â€¢ Bets {p1_card3[0]*100:.1f}% of the time")
    print(f"  â€¢ Checks {p1_card3[1]*100:.1f}% of the time")
    print(f"  â€¢ When facing bet: calls {p1_card3[4]*100:.1f}%, folds {p1_card3[3]*100:.1f}%")
    
    p2_card3 = final['player2_strategy'].iloc[:, 2]
    print(f"\nPlayer 2 with card 3 (middle):")
    print(f"  â€¢ After P1 checks: bets {p2_card3[0]*100:.1f}%, checks {p2_card3[1]*100:.1f}%")
    print(f"  â€¢ After P1 bets: calls {p2_card3[2]*100:.1f}%, folds {p2_card3[3]*100:.1f}%")
else:
    print("âš ï¸ No results to display. Run the full solver first!")
```

## ðŸŽ“ What You're Seeing

**Nash Equilibrium:**
- Neither player can improve by changing strategy alone
- Strategies are mathematically optimal
- Exploitability < 0.01 (less than 1 cent per $1 pot)

**Key Poker Concepts:**
- **Bluffing**: Betting with weak hands to fold out better hands
- **Value Betting**: Betting strong hands to get called by weaker hands
- **Bluff Catching**: Calling with medium hands against possible bluffs
- **Balance**: Mixing actions to remain unpredictable

**Applications Beyond Poker:**
- Negotiation strategies
- Competitive pricing
- Resource allocation
- Risk management
